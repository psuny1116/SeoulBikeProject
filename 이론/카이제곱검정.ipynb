```python
for i in range(1, 13):
    df_new_e = df_new[df_new['월'] == i]
    ad_cross = pd.crosstab(df_new_e['기존균형'], df_new_e['Queuing균형'])
    chi2, p, dof, ef = stats.chi2_contingency(ad_cross, correction = False)
    print(f'chi2  = {chi2:.2f}(df={dof:.2f}), p value > {p:.5f}.')
```
