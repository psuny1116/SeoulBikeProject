```python
# 여러개의 파일 병합

# 빈 데이터프레임 준비

raw = pd.DataFrame()

# 파일 하나씩 불러들여 합치기

for file in os.listdir('./공공자전거 대여소별 월별 이용정보'):
    # print(file)
    fpath = './공공자전거 대여소별 월별 이용정보/' + file
    # print(fpath)
    temp = pd.read_csv(fpath)
    raw = raw.append(temp, ignore_index = True)     # ignore_index = True는 기존 인덱스 무시
```

```python
raw.head()
```

```python
raw.tail()
```

```python
raw = raw.sort_values(by='대여일자')
raw
```

```python
raw.to_excel('./대여소별 월별 이용정보.xlsx', index = False)
```
