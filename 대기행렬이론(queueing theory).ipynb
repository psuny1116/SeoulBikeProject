# 대기행렬이론(queuing theory)

- 어떤 서비스를 받으려면 고객의 도착시간과 서비스시설의 서비스 시간이 다르게 되면 대기행렬이 생기거나 서비스시설이 휴지상태에 이르게 됨
- 대기행렬이론은 이와 같은 경우에 고객과 서비스시설과의 관계를 확률이론을 적용하여 모형을 작성하고 고객의 도착상황에 대응할 수 있는 경제적 규모를 결정하고자 하는 의사결정기법
- 고객의 도착 후 서비스 제공자의 입장에서 적절한 기다림의 수준을 유지하기 위한 서비스 능력 및 규모를 경제적으로 결정하고자 하는 이론
- 서비스 수준이 높아지면 서비스제공비용은 높아지나 고객 불만족비용은 증가
- 서비스 수준이 낮아지면 비용은 감소되나 고객불만족비용은 증가
- 따라서 이 두 가지의 상반된 비용의 최적점을 찾고자 함
- 서비스 시스템 내에서 서비스를 받으려는 고객의 도착시간과 서비스 시설의 이용시간이 다르기 때문에 발생함
<br>

- 대기행렬이 적용되는 분야: 서비스를 제공하는 창고가 몇 개 있으며, 여기에 서비스 받기를 원하는 고객이 도착하여 서비스를 받고 그곳을 떠나는 문제에 적용됨
    - 고속도로의 톨게이트
    - 은행 창구에서의 고객에 대한 서비스에 관련된 문제
    - 병원에서의 환자의 대기에 관련된 문제
    - 공장에서의 기계고장과 수리에 관련된 문제
    - 항구, 비행장의 시설규모 결정에 관련된 문제
    - 접객업소의 좌석수 결정에 관련된 문제

<br>
    
- 대기행렬과정의 구성요소 : 도착 - 대기행렬 - 서비스시설

<br>

- 대기행렬모형의 기본형태
    - 단일경로, 단일과정: 도착한 고객들이 한 줄로 서서 단일창구에서 한단계의 서비스를 받는 경우
    - 복수경로, 단일과정: 도착한 고객들이 여러 창구에서 한단계의 서비스를 받는 경우
    - 단일경로, 복수과정: 도착한 고객들이 한 줄로 서서 여러 단계의 서비스를 받는 경우
    - 복수경로, 복수과정: 도착한 고객들이 여러 줄로 서서 여러 단계의 서비스를 받는 경우

<br>

- 대기행렬이론의 가정
    
    1. 도착과 서비스시간의 분포: 도착분포단위시간당 서비스 시설에 도착하는 고객수는 포아송분포를 가정함<br>
        $ p_t(n) = \frac{(\lambda * t)^n}{n!} * e^(-\lambda*t) $<br>
        n: 도착의 수<br>
        $p_t(n)$: t시간에 도착의 수가 n이 될 확률<br>
        $\lambda$: 단위시간당 평균도착확률($\frac{1}{\lambda}$: 평균도착시간간격)<br>
        
        <br>
    
    2. 서비스시간의 분포서비스시간은 일반적으로 지수분포를 가정함<br>
        $ f(t) = (\mu * e)^(-\mu * t) $<br>
        t: 서비스시간<br>
        f(t): t시간에 따른 확률밀도<br>
        $\mu$: 평균서비스율($\frac{1}{\mu}$: 평균서비스시간)
        
        <br>
        
    3. 대기행렬규칙: 대기중 고객에게 서비스를 제고하는 순서를 결정하는 규칙으로서 선착순을 가정
    
    <br>
    
    4. 무한도착: 서비스를 받기 위해 도착하는 도착인원은 원칙적으로 무한모집단을 가정
    
    <br>
    
    5. 도착율과 서비스율: 서비스율(서비스가 완료되는 비율)이 고객의 도착율보다 크다고 가정(이는 그렇지 않으면 대기행렬은 계속 증가하기 때문) 
    
<br>

- 대기행렬이론의 목표: 대기비용과 서비스비용의 합계를 최소화시키는 것

    1. 대기비용: 서비스시설이 불충분할 때는 고객을 기다리게 함으로써 대기비용이 발생되며, 예로서 고객상실, 판매기회손실 등이 발생<br>
    $ E(C_w) = \bar{n_t} * C_w $ <br>
    $C_w$: 고객당 단위시간당 비용<br>
    $ E(C_w) $: 전체대기행렬 시스템내에서 기다리는 평균고객수($\bar{n_t}$)의 단위시간당 총비용
    
    <br>
    
    2. 서비스 비용: 서비스 시설을 증가시키면 대기비용은 감소시킬 수 있으나 그에 따른 서비스 비용이 증가함. 서비스비용의 예로서 시설에 대한 투자비, 추가적인 종업원 채용비 등<br>
    $ E(C_s) = S * C_s $ <br>
    $ E(C_s) $: 시간당 총서비스비용<br>
    $ C_s $: 서비스시설당 발생하는 단위시간당 비용<br>
    $ S $: 서비스시설의 수
    
    <br>
    
    3. 총대기비용: 대기행렬이론은 총대기비용을 최소화시키는 최적서비스 시설의 수를 결정하기 위해 개발되는 것<br>
    $ E(C_t) = E(C_w) + E(C_s) $

<br>

- 대기행렬모형의 성과측정(단일경로, 단일과정)
    - 단위시간당 평균적으로 도착하는 고객의 수(평균도착율) $\lambda$ $\to$ 고객의 평균도착시간 = $\frac{1}{\lambda}$
    - 단위시간당 평균적으로 서비스를 완료한 수(평균서비스율) $\mu$ $\to$ 고객의 평균서비스시간 = $\frac{1}{\mu}$

<br>
    
- 대기 중에 있는 고객의 평균수 $ L_q = \frac{\lambda^2}{\mu(\mu-\lambda)} $
- 평균 대기 시간 $ W_q = \frac{\lambda}{\mu(\mu-\lambda)} = \frac{L_q}{\lambda} $
- 시스템 내에 있는 고객의 평균체류시간 $ W = \frac{L}{\lambda} = W_q + \frac{1}{\mu} = \frac{1}{\mu-\lambda} $
- 시스템 내에 있는 고객의 평균 수 $ L = L_q + \frac{\lambda}{\mu} =  \frac{\lambda}{\mu-\lambda}$<br>
    - $\frac{\lambda}{\mu}$: 현재 서비스를 받고 있는 고객수
- 시스템 내에 있는 고객수가 n이 될 확률 $ P_{n} = (1 - \frac{\lambda}{\mu})(\frac{\lambda}{\mu})^n $
- 서비스 시설이 휴지가 될 확률 $ P_0 = (1-\frac{\lambda}{\mu}) = (1-\rho) $
- 서비스 이용률(utilization, $\rho$) $ \rho = \frac{\lambda}{\mu} $
    - 현재 서비스를 받고 있는 고객수
- 만일 $\rho$ >= 1이면 대기행렬이 무한대로 늘어나므로 $\rho$ < 1이 바람직. 즉, $\mu$ > $\lambda$이어야 함
- 시스템 내에 있는 고객의 수 (L) = 대기행렬 고객 수 + 서비스 중인 고객 수

#### 대기행렬 큐
- 큐: 먼저 들어온 데이터가 먼저 나가는 자료구조
- 선입선출
- 손님이 주문한 메뉴를 포스기에 입력하면 주방에서 새로운 주문서 출력
- 먼저 들어온 주문서부터 조리 시작
- 한 쉐프가 한 주문서의 메뉴 담당
- 시간대 별 평균 주문수가 높을수록 새로운 주문이 들어올 확률이 높음

#### 대기행렬이론과 Q-러닝 알고리즘을 적용한 지역문화축제 진입차량 주차분산 시뮬레이션 시스템 논문

- 대기 상황들과 서비스 되어지는 시간 등을 측정하기 위해 대기행렬 이론 적용
- 대기행렬이론에서의 주요변수는 도착특성, 서비스 특성, 그리고 서비스 위치에 관한 내용으로 구분됨(Marsanic et al., 2015)

서비스 특징|평균 서비스율(뮤$\mu$): 서비스 시간의 확률 분포, 동시에 서비스 할 수 있는 시설의 수
서비스 위치| FCFS: First come First served<br> LCFS : Last come First served<br> RSS: Random selection for service<br> PR: Priority<br> GD: General discipline

- 대기행렬이론에서는 대기행렬시스템에 접근하는 도착분포는 포아송분포(Poisson Distribution)를,
- 시스템의 서비스 시간 분포는 지수분포(Exponential Distribution)을 따른다고 가정
- 이 이론은 주어진 시간 내에 무작위적으로 발생하는 사건설명에 적합하며 차량 도착확률 등을 계산할 때 적용 됨
- 대기행렬이론에서 도착 확률은 $ p(x) =  \frac{\mu^x * e^-\mu}{x} $
- 위 식에서 x = 0, 1, 2 ... 이며 x는 일정시간 내에 도착하는 차량대수를 의미
- p(x)는 계수 기준 내에 x대 도착할 확률
- μ는 계수 기준 내 도착할 평균 차량 대수를 의미
- 모든 도착 실체(arrival entity)는 포아송 분포형태의 특별한 도착모델(arrival model)을 사용
- 대기행렬이론은 시간에 따른 순차적(sequence)처리가 가능하므로 이산적인 이벤트 시뮬레이션(discrete event simulation)을 가능하게 함
- 대기행렬이론의 큐(queue)는 서버의 대기시간과 유휴시간과 잘 동기화가 되어 있어야 함

#### 주차대기행렬 예측을 통한 기계식 주차장 적정 대기규모 산정에 관한 연구_석사논문
- 본 연구에서는 기계식 주차장의 특성과 대기행렬이론 등에 대해 고찰한 후 주차대기행렬에 영향을 미치는 인자인 평균서비스율과 평균도착률에 대하여 조사
- 현장조사 결과를 바탕으로 이론적 대기행렬 모형에 적용했으나 교통강도가 1을 초과할 경우 유의한 결과값을 얻지 못함
- 이에 때라 교통류 분석프로그램인 vissim 시뮬레이션을 단순 모델링하여 대기행렬을 분석을 수행하였고 분석 결과 규모별 기계식 주차장의 적정 대기 규모 산정을 위한 모델을 도출함
- 규모별 기계식 주차장의 주차대기행렬 분석을 통해 대기행렬을 고려한 기계식주차장의 적정 대기규모를 제시하는데 목적이 있음
<br>

- 서비스시간($t_n$): 대기행렬 변수인 평균서비스율 산정을 위해 기계식주차장 입출차 차량에 대한 평균서비스시간을 조사함
- 규모별 서비스시간($t_3$): 기계식 주차장의 규모별 입출차시간 분석을 위해 설치 업체를 대상으로 주차면수 규모별 입출차 시간을 조사함
- 입출차현황: 대기행렬 변수인 평균도착률 산정을 위해 기계식주차장의 피크시간 동안 입출차 현황을 분석하여 기계식 주차장 규모별 평균도착률을 결정함
- 조사결과를 바탕으로 기계식 주차장의 평균도착률과 평균서비스시간을 구하고 기계식 주차장 규모별 대기행렬 예측모델을 개발 및 기계식 주차장의 적정 대기규모를 제시함
<br>

- 대기행렬시스템 구성요소
    1. 대상고객<br>
        - 고객의 규모
            - 유한고객: 공장에서 가동중인 설비, 전산실의 pc등
            - 무한고객: 백화점의 고객, 주차장 고객 등<br>
        - 고객의 도착유형<br>
            - 고객의 도착간격시간 분포
            - 한번에 도착할 수 있는 고객의 크기
            - 고객이 어떠한 규칙이나 패턴을 가지고 도착하는가 등<br>
        - 고객의 태도
            - 시스템에 도착한 고객이 인내심을 가지고 있느냐 없느냐 등의 문제
    2. 도착분포
        - 포아송분포
            - 불규칙한 고객의 도착을 표현해 주는 이산확률분포
            - 평균도착율이 $\lambda$일때, t시간내에 n 고객이 도착할 확률
            - $ p_n(t) = \frac{e^-x*\lambda*t^n}{n}$, n = 0, 1, 2 ...
        - 지수분포
            - 포아송 분포에서 도착하는 고객들의 시간간격을 나타내 주는 연속확률분포
    3. 대기행렬의 길이
        - 대기행렬의 허용길이가 유한인지 무한인지의 문제
    4. 서비스시설
        - 단일채널-단일단계 대기시스템
        - 다중채널-단일단계 대기시스템
    5. 서비스 정책
        - 선입선출(FCFS)
        - 후입선출(LCFS)
        - 무작위 규칙(SIRO)
        - 우선순위에 의한 규칙
            - 선점규칙(preemptive discipline): 서비스 중인 고객의 서비스 중단
            - 비선점규칙(non preemptive discipline): 서비스 중인 고객의 서비스는 완료
    6. 서비스 시간의 분포
        - 지수분포
        - 평균서비스 시간이 $\frac{1}{\mu}$라 하면, 서비스 시간 t의 확률밀도함수는 $f(t)=\mu*e^(-\mu*t)$, t>=0
        - f(t)의 누적분포함수 F(T) = 서비스시간이 T보다 작을 확률<br>
            $p_r(t<=T) = \int_0^tf(t)dt = 1 - e^(-\mu*T)$, T>=0
            
<br>

- 대기행렬 모형
    1. 대기행렬모형의 표시방법
        - (a/b/c) : (d/e/f)
            - a : 고객의 도착빈도 분포
            - b : 서비스시간 분포
            - c : 서비스 시설의 수
            - d : 서비스규칙
            - e : 시스템내의 최대허용고객수(서비스 받는 고객+대기행렬의 고객)
            - f : 고객 모집단의 크기
        - a, b에 사용되는 기호와 내용
            - M : 도착분포가 포아송분포이거나 서비스 시간 분포가 지수분포인 경우
            - D : 도착분포나 서비스 시간이 확정적인 경우
            - $E_k$ : 도착분포나 서비스시간 분포가 감마분포인 경우
            - G : 도착분포나 서비스시간 분포가 기타의 일반분포를 따르는 경우
        - 서비스규칙을 나타내는 기호 d 의 표시
            - FCFS : 선입선출(first come, first service)
            - LCFS : 후입선출(last come, first service)
            - SIRO : 무작위 순서(service in random order)
            - GD : 일반적 규칙(general dicipline)
            - PRP : 우선순위 규칙(priority rule)
