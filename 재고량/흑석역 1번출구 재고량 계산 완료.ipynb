```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'


pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_excel('./1년 365일 8시 재고량.xlsx')
df
```

```python
df = df[(df['대여소명'] == '흑석역 1번출구')].reset_index(drop = True)
df
```

```python
data = pd.read_csv('./1년 1시간별 대여량, 반납량.csv', encoding = 'cp949')
data
```

```python
data = data[(data['대여소명'] == '흑석역 1번출구')].reset_index(drop = True)
data
```

```python
df_new = pd.merge(data, df, on = ['자치구', '대여소명', '대여소번호', '년', '월', '일'], how = 'outer')
df_new.head()
```

```python
df_new = df_new.drop(['시간_y'], axis = 1)
df_new = df_new.rename(columns = {'시간_x' : '시간'})
df_new
```

```python
df_new.fillna(0, inplace = True)
df_new
```

```python
a = df_new[df_new['시간'] == 8]
a
```

```python
b = df_new[df_new['시간'] != 8]
b['초기재고량'] = 0
b
```

```python
df_new_z = pd.merge(b, a, on = ['자치구', '대여소명', '대여소번호', '년', '월', '일','시간', 'c', 'return_count', 'rent_count', '거치대수', '초기재고량'], how = 'outer')
df_new_z.head()
```

```python
df_new_z = df_new_z.sort_values(['자치구', '대여소명', '년', '월', '일', '시간'], ignore_index = True)
df_new_z = df_new_z.astype({'시간' : 'int64', 'c' : 'int64'})
df_new_z.head()
```

```python
df_new_z['재고량'] = df_new_z['초기재고량'] + df_new_z['return_count'] - df_new_z['rent_count']
df_new_z.head(100)
```

```python
for i in range(1, len(df_new_z)):
    if df_new_z['시간'][i] == 8:
        df_new_z['재고량'][i] = df_new_z['초기재고량'][i]
    else:
        df_new_z['재고량'][i] += df_new_z['재고량'][i - 1]
df_new_z.head(100)
```

```python
df_new_z = df_new_z.drop(['대여소번호', 'c', 'return_count', 'rent_count', '거치대수', '초기재고량'], axis = 1)
df_new_z
```

```python
sol = df_new_z.groupby(['자치구', '대여소명', '년', '월', '시간'])['재고량'].sum().reset_index()
sol
```

```python
sol['재고량'] = sol['재고량'] / 30
sol
```

```python
answer = sol.copy()
```

```python
answer['3사분위수'] = np.percentile(answer['재고량'],75)
answer['1사분위수'] = np.percentile(answer['재고량'],25)
answer
```

```python
answer['균형'] = '균형'
for i in range(len(answer)):
    if answer['3사분위수'][i]>=answer['재고량'][i] >= answer['1사분위수'][i]:
        answer['균형'][i] = '균형'
    else:
        answer['균형'][i] = '불균형'
answer
```

```python
answer.to_csv('./흑석역 1번출구 균형, 불균형.csv', index = False, encoding = 'cp949')
```
