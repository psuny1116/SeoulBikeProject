```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'


pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
df1 = pd.read_csv('./일별 대여소별 자전거 거치수량(2020년).csv', encoding = 'cp949')
df2 = pd.read_csv('./일별 대여소별 자전거 거치수량(21년1~5월).csv', encoding = 'cp949')
df = pd.merge(df1, df2, on = ['일시', '대여소번호', '대여소ID', '거치수량'],how='outer')
df
```

```python
station = pd.read_csv('./대학원/데이터/정리 데이터/공공자전거 대여소 정보(21.06월 기준).csv')
station.head()
```

```python
station = station.fillna(0)
```

```python
station['거치대수'] = station['LCD 거치대수'] + station['QR 거치대수']
station.head()
```

```python
station = station.rename(columns = {'대여소 번호' : '대여소번호', '보관소(대여소)명' : '대여소명'})
station
```

```python
df_new = pd.merge(df, station[['자치구', '대여소번호', '대여소명', '거치대수']], on = '대여소번호')
df_new.head()
```

```python
df_new['일시'] = pd.to_datetime(df_new['일시'])
df_new['년'] = df_new['일시'].dt.year
df_new['월'] = df_new['일시'].dt.month
df_new['일'] = df_new['일시'].dt.day
df_new['시간'] = df_new['일시'].dt.hour
```

```python
df_new.head()
```

```python
df_new = df_new.drop(['대여소ID', '거치대수'], axis = 1)
df_new.head()
```

```python
df_new = df_new[['일시', '대여소명', '대여소번호', '년', '월', '일', '시간', '거치수량']]
df_new.head()
```

```python
df_new = df_new.rename(columns = {'거치수량' : '재고량'})
df_new.head()
```

```python
df_new = df_new[df_new['대여소명'] == '롯데월드타워(잠실역2번출구 쪽)']
df_new.reset_index(drop = True)
```

```python
df_new.to_excel('./롯데월드타워(잠실역2번출구 쪽) 1년 365일 8시 재고량.xlsx', index = False)
```
