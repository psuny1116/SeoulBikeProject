```python
df = pd.read_csv('./1년 1시간별 Ls, Ws.csv', encoding = 'cp949')
df
```

```python
df = df[df['대여소명'] == '현대고등학교 건너편']
df
```

```python
df = df.drop(['대여소번호', 'lambda', 'mu', 'c', 'rho', 'p_c', 'p_1+'], axis = 1)
df
```

```python
data = pd.read_csv('./현대고등학교 건너편 균형, 불균형.csv', encoding = 'cp949')
data
```

```python
data = data.drop(['3사분위수', '1사분위수'], axis = 1)
data
```

```python
df_new = pd.merge(df,data, on = ['자치구', '대여소명','년', '월', '시간'],how='outer')
df_new
```

```python
len(df_new)
```

```python
# 재고량이 ls보다 많거나 같을 경우 불균형 아님 = 균형임

z = df_new[df_new['L_s'] <= df_new['재고량']]
z
```

```python
len(z)
```

```python
# TP

len(z[z['균형'] == '균형'])
```

```python
# FN

len(z[z['균형'] == '불균형'])
```

- 재고량이 ls보다 많거나 같을 경우 균형인데 불균형이라고 판단한 수가 111개

- 균형인데 균형이라고 판단한 수가 143개

```python
# ls보다 재고량이 작으므로 불균형

w = df_new[df_new['L_s'] > df_new['재고량']]
w
```

```python
len(w)
```

```python
# FP

len(w[w['균형'] == '균형'])
```

```python
# TN

len(w[w['균형'] == '불균형'])
```

- 불균형인데 균형이라고 판단한 수는 1
- 불균형인데 불균형이라고 판단한 수는 33

# confusion matrix

0 |균형|불균형
-|-|-
균형|TP|FP
불균형|FN|TN


1. 정확도 accuracy = 실제(내) 데이터와 예측(기존방법)데이터가 얼마나 같은지를 평가하는 지표 = TP + TN / TP+TN+FP+FN
2. 정밀도 precision = 균형으로 예측한 데이터 중 실제로 균형인 데이터 비율 = TP / TP+FP
3. 재현율 recall = 실제 균형 중 균형으로 예측한 비율 = TP / TP+FN

=> 정밀도에 초점. (불균형을 균형으로 예측하여 재배치하지 않으면 서비스에 차질)


```python
# 정확도

(len(z[z['균형'] == '균형']) + len(w[w['균형'] == '불균형'])) / len(df_new)
```

```python
# 정밀도

len(z[z['균형'] == '균형']) / (len(z[z['균형'] == '균형']) + len(w[w['균형'] == '균형']))
```

```python
# 재현율

len(z[z['균형'] == '균형']) / (len(z[z['균형'] == '균형']) + len(z[z['균형'] == '불균형']))
```
