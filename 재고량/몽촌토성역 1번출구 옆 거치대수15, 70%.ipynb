## L_s가 Q1보다 작거나, Q3보다 크면 불균형

### L_s가 q1보다 작지만 이때 p_1+이 충분히 크다면 균형
- ls가 q1보다 작고 p_1+가 0.5보다 작으면 불균형

```python
# ls가 q1보다 작으므로 불균형

e = df_new[(df_new['L_s'] < df_new['1사분위수']) & (df_new['p_1+'] < 0.6)]
e
```

```python
for i in range(len(df_new)):
    if (df_new['L_s'][i] < df_new['1사분위수'][i]) & (df_new['p_1+'][i] < 0.6):
        df_new['내 균형'][i] = '불균형'
```

```python
# ls가 q3보다 크므로 불균형

r = df_new[(df_new['L_s'] > df_new['3사분위수']) & (df_new['L_s'] > (15*0.7))]
r
```

```python
for i in range(len(df_new)):
    if (df_new['L_s'][i] > df_new['3사분위수'][i]) & (df_new['L_s'][i] > (15*0.7)):
        df_new['내 균형'][i] = '불균형'
```

# confusion matrix

0 |균형|불균형
-|-|-
균형|TP|FP
불균형|FN|TN


1. 정확도 accuracy = 실제(내) 데이터와 예측(기존방법)데이터가 얼마나 같은지를 평가하는 지표 = TP + TN / TP+TN+FP+FN
2. 정밀도 precision = 균형으로 예측한 데이터 중 실제로 균형인 데이터 비율 = TP / TP+FP
3. 재현율 recall = 실제 균형 중 균형으로 예측한 비율 = TP / TP+FN

=> 정밀도에 초점. (불균형을 균형으로 예측하여 재배치하지 않으면 서비스에 차질)


```python
# 정확도

(len(z[z['균형'] == '균형']) + len(w[w['균형'] == '불균형'])) / len(df_new)
```

```python
# 정밀도

len(z[z['균형'] == '균형']) / (len(z[z['균형'] == '균형']) + len(w[w['균형'] == '균형']))
```

```python
# 재현율

len(z[z['균형'] == '균형']) / (len(z[z['균형'] == '균형']) + len(z[z['균형'] == '불균형']))
```
