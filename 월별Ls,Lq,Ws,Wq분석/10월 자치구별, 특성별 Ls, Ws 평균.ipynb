```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'


pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_excel('./data/10월 대여소별 Ls, Lq, Ws, Wq.xlsx')
df
```

```python
df = df.drop(['lambda_15', 'mu_15', 'rho', 'p_c', 'L_q', 'W_q'], axis = 1)
df.head()
```

```python
df_ls = df.pivot_table(index = ['자치구'], values = ['L_s']).reset_index()
df_ls
```

```python
df_ws = df.pivot_table(index = ['자치구'], values = ['W_s']).reset_index()
df_ws
```

```python
df_new = pd.merge(df_ls, df_ws, on = ['자치구'], how='outer')
df_new
```

```python
df_new.to_excel('./10월 자치구별 Ls, Ws 평균.xlsx', index = False)
```

```python
s = pd.DataFrame(df['대여소명'].unique()).rename(columns = {0 : 'station'})

school = []
subway = []
apart = []
park = []
for i in range(len(s)):
    if '출구' in s['station'][i]:
        subway.append(s['station'][i])
    elif '학교' in s['station'][i]:
        school.append(s['station'][i])
    elif '아파트' in s['station'][i]:
        apart.append(s['station'][i]) 
    elif '공원' in s['station'][i]:
        park.append(s['station'][i])
```

```python
school_data = []
for i in range(len(df)):
    if df['대여소명'][i] in school:
        school_data.append(df.loc[i])
school_data = pd.DataFrame(school_data)
school_data['특성'] = '학교'

subway_data = []
for i in range(len(df)):
    if df['대여소명'][i] in subway:
        subway_data.append(df.loc[i])
subway_data = pd.DataFrame(subway_data)
subway_data['특성'] = '지하철'

apart_data = []
for i in range(len(df)):
    if df['대여소명'][i] in apart:
        apart_data.append(df.loc[i])
apart_data = pd.DataFrame(apart_data)
apart_data['특성'] = '아파트'

park_data = []
for i in range(len(df)):
    if df['대여소명'][i] in park:
        park_data.append(df.loc[i])
park_data = pd.DataFrame(park_data)
park_data['특성'] = '공원'
```

```python
data = pd.merge(school_data, subway_data, on=['특성', '자치구', '대여소명', '시간', 'L_s', 'W_s'], how = 'outer')
data = pd.merge(data, apart_data, on=['특성', '자치구', '대여소명', '시간', 'L_s', 'W_s'], how = 'outer')
data = pd.merge(data, park_data, on=['특성', '자치구', '대여소명', '시간', 'L_s', 'W_s'], how = 'outer')
data
```

```python
data_ls = data.pivot_table(index = ['특성'], values = ['L_s']).reset_index()
data_ls
```

```python
data_ws = data.pivot_table(index = ['특성'], values = ['W_s']).reset_index()
data_ws
```

```python
data_new = pd.merge(data_ls, data_ws, on = ['특성'], how='outer')
data_new
```

```python
data_new.to_excel('./10월 특성별 Ls, Ws 평균.xlsx', index = False)
```
