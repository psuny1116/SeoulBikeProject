```python
df_rental = df[['대여일시', '대여대여소명','대여대여소번호', '대여년', '대여월', '대여일', '대여요일', '대여시간', '대여시간_분']]
df_rental.reset_index(drop = True, inplace = True)
df_rental.head()
```

```python
df_rental2 = df_rental.groupby(['대여대여소명','대여대여소번호', '대여년', '대여월', '대여일', '대여시간']).size().reset_index().rename(columns = {0 : 'rent_count'})
df_rental2
```

```python
df_rental2['mu_15'] = df_rental2['rent_count'] # 1시간단위
df_rental2
```

```python
df_return = df[['반납일시', '반납대여소명', '반납대여소번호','반납년', '반납월', '반납일', '반납요일', '반납시간', '반납시간_분']]
df_return.reset_index(drop = True, inplace = True)
df_return
```

```python
df_c = pd.DataFrame(df_return[['반납대여소명','반납대여소번호','반납년', '반납월', '반납일', '반납시간']].value_counts()).rename(columns = {0 : 'return_count'}).reset_index()
df_c.reset_index(drop=True, inplace = True)
df_c['lambda15'] = df_c['return_count']     # 1시간단위
df_c.columns = ['대여소명','대여소번호','년', '월','일', '시간','return_count','lambda15']
df_c
```

```python
return_ubound = pd.DataFrame(df_c.groupby(['대여소명','대여소번호'])['return_count'].agg(lambda x : np.floor(np.percentile(x,75)+(np.percentile(x,75)-np.percentile(x,25))*1.5))).reset_index().rename(columns = {'return_count' : 'upperbound'})
return_ubound.head()
```

```python
df['rho'] = float(-1)

for i in range(len(df)):
    if df['mu_15'][i] == 0:
        df['rho'][i] = 0
    else:
        df['rho'][i] = df['lambda_15'][i] / df['mu_15'][i]
df
```

```python
# p_0: 대여소에 자전거가 0대 있을 확률

df['p_0'] = float(-1)

for i in range(len(df)):
    if df['rho'][i] == 1:
        df['p_0'][i] = 1 / (1 + df['c'][i]) 
    else:
        df['p_0'][i] = (df['rho'][i]**0) * ((1 - df['rho'][i]) / (1 - (df['rho'][i]**(df['c'][i] + 1))))
df
```

```python
# L_s : 대여소 내 평균 자전거 수

df['L_s'] = float(-1)

for i in range(len(df)):
    rho = df['rho'][i]
    c = df['c'][i]
    if df['rho'][i] == 1:
        df['L_s'][i] = c / 2
    else:
        df['L_s'][i] = (rho / (1 - rho)) * ((1 - ((c + 1) * rho**c) + (c * rho**(c + 1))) / (1 - (rho**(c + 1))))
df
```
